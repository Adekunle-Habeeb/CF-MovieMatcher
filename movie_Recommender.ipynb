```python
merged_df = pd.merge(ratings, movies, left_on='movieId', right_on='id', how='inner')
```

```python
merged_df = merged_df.drop(['index', 'budget', 'homepage', 'id', 'keywords','original_language', 'original_title', 'overview', 'popularity', 'production_companies', 'production_countries', 'release_date',
       'revenue', 'runtime', 'spoken_languages', 'status', 'tagline',
       'vote_average', 'vote_count', 'cast', 'crew', 'director', 'timestamp'], axis=1)
```

```python
merged_df
```

```python
merged_df.isnull().sum()
```

```python
user_ratings = merged_df.pivot_table(index='userId', columns='title', values='rating')
```

```python
movie_genres = merged_df.groupby('title')['genres'].first().reset_index()
```

```python
user_ratings = user_ratings.dropna(thresh=2, axis=1).fillna(0)
```

```python
user_ratings
```

```python
item_similarity_df = user_ratings.corr(method='pearson')
item_similarity_df
```

```python
item_similarity_df.to_csv('item_similarity.csv', index=False)
```

```python
def get_similar_movies(movie_name, user_rating):
    if movie_name not in item_similarity_df.index:
        print(f"Movie '{movie_name}' not found in the similarity matrix.")
        return pd.Series()
    similar_score = item_similarity_df.loc[movie_name] * user_rating  
    return similar_score.sort_values(ascending=False)
```

```python
action_lover = [('You Only Live Twice', 2), ('28 Days Later', 3), ('Zodiac', 5)]
similar_movies = pd.DataFrame()

input_genres = set()
for movie, rating in action_lover:
    similar = get_similar_movies(movie, rating)
    similar = similar.to_frame(name='score')
    similar['title'] = similar.index
    similar_movies = pd.concat([similar_movies, similar], ignore_index=True)
    
    
    movie_genre = movie_genres[movie_genres['title'] == movie]['genres'].iloc[0] if not movie_genres[movie_genres['title'] == movie].empty else ''
    input_genres.update(movie_genre.split('|'))
```

```python
movie_genres.to_csv('movie_genres.csv', index=False)
```

```python
recommended_movies = similar_movies.groupby('title')['score'].sum().sort_values(ascending=False)
recommended_movies = recommended_movies.to_frame().reset_index()

# Merge with movie_genres to get the genres
recommended_movies_with_genres = recommended_movies.merge(movie_genres, on='title', how='left')

# Filter to include only movies that share at least one genre with input movies
recommended_movies_with_genres['genres_list'] = recommended_movies_with_genres['genres'].str.split('|')
recommended_movies_with_genres = recommended_movies_with_genres[recommended_movies_with_genres['genres_list'].apply(lambda x: bool(set(x) & input_genres))]

# Reorder and format
recommended_movies_with_genres = recommended_movies_with_genres[['title', 'genres', 'score']]
recommended_movies_with_genres['score'] = recommended_movies_with_genres['score'].round(2)

print(recommended_movies_with_genres.head(10).to_string(index=False))
```

```python
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise import accuracy
```

```python
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(merged_df[['userId', 'title', 'rating']], reader)
```

```python
trainset, testset = train_test_split(data, test_size=0.25, random_state=42)
```

```python
model = SVD(n_factors=100, n_epochs=20, lr_all=0.005, reg_all=0.02)
model.fit(trainset)
```

```python
predictions = model.test(testset)
```

```python
rmse = accuracy.rmse(predictions)
mae = accuracy.mae(predictions)

print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
```

```python
def get_top_n_recommendations(model, user_id, n=10):
    # Get all movies the user hasn't rated
    user_movies = set(merged_df[merged_df['userId'] == user_id]['title'])
    all_movies = set(merged_df['title'])
    movies_to_predict = all_movies - user_movies

    # Make predictions for all these movies
    predictions = [model.predict(user_id, movie) for movie in movies_to_predict]

    # Sort predictions by estimated rating
    top_n = sorted(predictions, key=lambda x: x.est, reverse=True)[:n]

    return [(pred.iid, pred.est) for pred in top_n]

# Example usage
user_id = 1  # replace with an actual user ID from your dataset
recommendations = get_top_n_recommendations(model, user_id)

print(f"Top 10 recommendations for user {user_id}:")
for movie, rating in recommendations:
    print(f"{movie}: Estimated rating = {rating:.2f}")

# Add genres to recommendations
recommended_movies_with_genres = pd.DataFrame(recommendations, columns=['title', 'estimated_rating'])
recommended_movies_with_genres = recommended_movies_with_genres.merge(movie_genres, on='title', how='left')
print(recommended_movies_with_genres[['title', 'genres', 'estimated_rating']])
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def get_movie_list(n=50):
    # Randomly select n movies from your movie_genres DataFrame
    return movie_genres.sample(n=n)['title'].tolist()

def get_user_ratings(movie_list):
    user_ratings = {}
    print("Please rate at least 3 movies (1-5 stars). Type 'done' when finished.")
    while len(user_ratings) < 3 or input("Rate another movie? (y/n): ").lower() != 'n':
        movie = input(f"Choose a movie from the list {movie_list}: ")
        if movie in movie_list:
            rating = float(input(f"Rate {movie} (1-5 stars): "))
            user_ratings[movie] = rating
        else:
            print("Movie not in list. Please choose again.")
    return user_ratings

def get_recommendations(user_ratings, item_similarity_df, movie_genres, n=10):
    # Create a user vector
    user_vector = pd.Series(user_ratings)
    user_vector = user_vector.reindex(item_similarity_df.index, fill_value=0)
    
    # Calculate weighted sum of similarities
    weighted_sums = item_similarity_df.dot(user_vector)
    
    # Sort and get top N recommendations
    recommendations = weighted_sums.sort_values(ascending=False)
    recommendations = recommendations[~recommendations.index.isin(user_ratings.keys())]
    top_n = recommendations.head(n)
    
    # Add genres to recommendations
    rec_with_genres = pd.DataFrame({'title': top_n.index, 'score': top_n.values})
    rec_with_genres = rec_with_genres.merge(movie_genres[['title', 'genres']], on='title', how='left')
    
    return rec_with_genres
```

```python
movie_list = get_movie_list(50)

# Get user ratings
user_ratings = get_user_ratings(movie_list)

# Get recommendations
recommendations = get_recommendations(user_ratings, item_similarity_df, movie_genres)

# Print recommendations
print("\nTop 10 Recommendations:")
print(recommendations.to_string(index=False))
```
